```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
test_sel_70 = pd.read_csv("/workspaces/ml-proyect-decision-trees/data/processed/test_x_sel_70.csv")
train_sel_70 = pd.read_csv("/workspaces/ml-proyect-decision-trees/data/processed/train_x_sel_70.csv")

test_sel_80 = pd.read_csv("/workspaces/ml-proyect-decision-trees/data/processed/test_x_sel_80.csv")
train_sel_80 = pd.read_csv("/workspaces/ml-proyect-decision-trees/data/processed/train_x_sel_80.csv")
```

70% best

```python
train_sel_70.head()
```

```python
X_train = train_sel_70.drop("Outcome", axis=1)
y_train = train_sel_70["Outcome"]

X_test = test_sel_70.drop("Outcome", axis=1)
y_test = test_sel_70["Outcome"]

X_train.head(), y_train.head()
```

```python
model = DecisionTreeClassifier(random_state = 24)
model.fit(X_train, y_train)
```

```python
fig = plt.figure(figsize=(15,15))

tree.plot_tree(model, feature_names = list(X_train.columns), class_names = ["0", "1"], filled = True)

plt.show()
```

```python
y_pred = model.predict(X_train)
y_pred
```

```python
accuracy_score(y_train, y_pred)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
# We define the parameters by hand that we want to adjust
hyperparams = {
    "criterion": ["gini", "entropy", "log_loss", None],
    "splitter": ["best", "random", None],
    "max_depth": np.append(np.arange(3, 30, 4), None),
    "min_samples_split": np.arange(2, 11, 4),
    "min_samples_leaf": np.arange(2, 150, 10),
    "max_features": ["auto", "sqrt", "log2"],
    "random_state": [24]
}

# We initialize the grid
grid = GridSearchCV(model, hyperparams, scoring = "accuracy", cv = 5)
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
best_model = grid.best_estimator_
```

```python
fig = plt.figure(figsize=(15,15))

tree.plot_tree(best_model, feature_names = list(X_train.columns), class_names = ["0", "1"], filled = True)

plt.show()
```

```python
y_pred = best_model.predict(X_train)
y_pred
```

```python
accuracy_score(y_train, y_pred)
```

```python
y_pred = best_model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
